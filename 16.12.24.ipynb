```python
# Ben√∂tigte Bibliotheken importieren
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --- PANDAS BEISPIELE ---
# 1. DataFrame erstellen und grundlegende Operationen
print("### Pandas Beispiele ###")
# Beispieldaten erstellen
verkaufsdaten = {
    'Produkt': ['Laptop', 'Maus', 'Tastatur', 'Monitor', 'Laptop'],
    'Preis': [1200, 25, 45, 300, 1300],
    'Datum': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02', '2024-01-03']
}
df = pd.DataFrame(verkaufsdaten)

# Grundlegende Analysen
print("Durchschnittliche Preise:")
print(df.groupby('Produkt')['Preis'].mean())

# Datums-Analyse
df['Datum'] = pd.to_datetime(df['Datum'])
print("\nVerk√§ufe pro Tag:")
print(df.groupby('Datum').size())

# --- NUMPY BEISPIELE ---
print("\n### NumPy Beispiele ###")
# Array-Operationen
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([2, 4, 6, 8, 10])

print("Array Mathematik:")
print("Summe:", arr1 + arr2)
print("Produkt:", arr1 * arr2)
print("Quadrat:", np.square(arr1))

# Matrix-Operationen
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

print("\nMatrix Multiplikation:")
print(np.dot(matrix1, matrix2))

# Zufallszahlen
zufallszahlen = np.random.normal(100, 10, 1000)  # Mittelwert 100, Standardabweichung 10
print("\nStatistik der Zufallszahlen:")
print(f"Mittelwert: {np.mean(zufallszahlen):.2f}")
print(f"Standardabweichung: {np.std(zufallszahlen):.2f}")

# --- SCIPY BEISPIELE ---
print("\n### SciPy Beispiele ###")
# Statistische Tests
gruppe1 = np.random.normal(100, 15, 100)  # Kontrollgruppe
gruppe2 = np.random.normal(105, 15, 100)  # Testgruppe

# T-Test durchf√ºhren
t_stat, p_value = stats.ttest_ind(gruppe1, gruppe2)
print("T-Test Ergebnisse:")
print(f"t-Statistik: {t_stat:.4f}")
print(f"p-Wert: {p_value:.4f}")

# Korrelationsanalyse
x = np.random.normal(0, 1, 1000)
y = 2 * x + np.random.normal(0, 0.5, 1000)  # Lineare Beziehung mit Rauschen
korrelation = stats.pearsonr(x, y)
print("\nKorrelationsanalyse:")
print(f"Korrelationskoeffizient: {korrelation[0]:.4f}")
print(f"p-Wert: {korrelation[1]:.4f}")

# Visualisierung der Daten (mit Pandas und Matplotlib)
plt.figure(figsize=(12, 4))

# Plot 1: Verk√§ufe pro Produkt
plt.subplot(131)
df.groupby('Produkt')['Preis'].mean().plot(kind='bar')
plt.title('Durchschnittspreis pro Produkt')
plt.xticks(rotation=45)

# Plot 2: Normalverteilte Zufallszahlen
plt.subplot(132)
plt.hist(zufallszahlen, bins=30)
plt.title('Verteilung der Zufallszahlen')

# Plot 3: Korrelation
plt.subplot(133)
plt.scatter(x, y, alpha=0.5)
plt.title('Korrelationsplot')

plt.tight_layout()
plt.show()

# Erweiterte Pandas-Funktionen
# 1. Pivot-Tabelle
pivot_tabelle = df.pivot_table(
    values='Preis',
    index='Datum',
    columns='Produkt',
    aggfunc='sum'
)
print("\nPivot-Tabelle der Verk√§ufe:")
print(pivot_tabelle)

# 2. Rollierender Durchschnitt (wenn mehr Daten vorhanden w√§ren)
df_erweitert = pd.DataFrame({
    'Datum': pd.date_range(start='2024-01-01', periods=10),
    'Wert': np.random.normal(100, 10, 10)
})
print("\nRollierender Durchschnitt:")
print(df_erweitert.set_index('Datum')['Wert'].rolling(window=3).mean())
```

```python
from scipy import stats
import numpy as np

# Zufallsdaten erstellen
daten1 = np.random.normal(100, 15, 100)  # Normalverteilung
daten2 = np.random.normal(105, 15, 100)  # Leicht verschobene Normalverteilung

# Statistischen Test durchf√ºhren
t_stat, p_wert = stats.ttest_ind(daten1, daten2)

print("T-Test Ergebnisse:")
print(f"t-Statistik: {t_stat:.4f}")
print(f"p-Wert: {p_wert:.4f}")
```

```python
import yfinance as yf
import plotly.graph_objects as go

# Bitcoin-Daten laden (BTC-USD ist der Ticker f√ºr Bitcoin)
btc = yf.Ticker("XRP-USD")
daten = btc.history(period="1y")

# Interaktiven Candlestick-Chart erstellen
fig = go.Figure(data=[go.Candlestick(x=daten.index,
                open=daten['Open'],
                high=daten['High'],
                low=daten['Low'],
                close=daten['Close'])])

# Layout anpassen und mehr Details hinzuf√ºgen
fig.update_layout(
    title='XRP Kurs (XRP-USD)',
    yaxis_title='Preis in USD',
    xaxis_title='Datum',
    template='plotly_dark'  # Dunkles Theme
)

# Volumen als Balken hinzuf√ºgen
fig.add_trace(go.Bar(
    x=daten.index,
    y=daten['Volume'],
    name='Volume',
    yaxis='y2'
))

# Zweite Y-Achse f√ºr Volumen
fig.update_layout(
    yaxis2=dict(
        title='Volumen',
        overlaying='y',
        side='right'
    )
)

# Plot anzeigen
fig.show()
```
